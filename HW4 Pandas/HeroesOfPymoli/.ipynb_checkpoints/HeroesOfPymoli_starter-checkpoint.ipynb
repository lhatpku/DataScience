### Heroes Of Pymoli Data Analysis
* Of the 1163 active players, the vast majority are male (84%). There also exists, a smaller, but notable proportion of female players (14%).

* Our peak age demographic falls between 20-24 (44.8%) with secondary groups falling between 15-19 (18.60%) and 25-29 (13.4%).

```python
# Dependencies and Setup
import pandas as pd
import numpy as np
import locale
locale.setlocale(locale.LC_ALL, '')

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()
```

## Player Count

* Display the total number of players


```python
player_count_df = pd.DataFrame([{"Total Players":len(purchase_data["SN"].unique())}])
player_count_df
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
# Stat Calculate
unique_items_count = len(purchase_data["Item ID"].unique())
average_price = purchase_data["Price"].mean()
purchases_count = purchase_data["Price"].count()
total_revenue = purchase_data["Price"].sum()

# Create DataFrame to hold results
purchase_analysis_df = pd.DataFrame([{"Number of Unique Items": unique_items_count,
                              "Average Price": locale.currency(average_price),
                              "Number of Purchases": purchases_count,
                               "Total Revenue": locale.currency(total_revenue)}])

purchase_analysis_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
# Count Number Calculate
male_player_count = int(purchase_data.drop_duplicates(subset=["SN"])["Gender"].value_counts()["Male"])
female_player_count = int(purchase_data.drop_duplicates(subset=["SN"])["Gender"].value_counts()["Female"])
total_player_count = int(purchase_data.drop_duplicates(subset=["SN"])["Gender"].count())
other_player_count = total_player_count - female_player_count - male_player_count

# Statistical Data Frame
gender_demographics_analysis_data = np.array([[male_player_count,round(male_player_count/total_player_count*100,2)],
                                             [female_player_count,round(female_player_count/total_player_count*100,2)],
                                             [other_player_count,round(other_player_count/total_player_count*100,2)]])

gender_demographics_analysis_df = pd.DataFrame(gender_demographics_analysis_data,index=["Male","Female","Other/Non-Discloased"],
                                              columns = ["Total Count","Percentage of Players"])

gender_demographics_analysis_df
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# Calc stats for each column
purchase_data_by_gender = purchase_data.groupby(["Gender"])
purchase_count = round(purchase_data_by_gender["Gender"].count(),2)
avg_purchase_price = purchase_data_by_gender["Price"].mean()
total_purchase_value = purchase_count * avg_purchase_price
purchase_count_unique = purchase_data.drop_duplicates(subset=["SN"]).groupby(["Gender"])["Gender"].count()
avg_purchase_per_person = round(total_purchase_value / purchase_count_unique,2)

# Aggregate each field into data frame
purchase_analysis_df = pd.concat([purchase_count,avg_purchase_price,total_purchase_value,avg_purchase_per_person],axis=1)
purchase_analysis_df.columns = ["Purchase Count","Average Purchase Price","Total Purchase Value","Avg Total Purchase per Person"]

# Change to dollar format
purchase_analysis_df["Average Purchase Price"] = purchase_analysis_df["Average Purchase Price"].map("${:.2f}".format)
purchase_analysis_df["Total Purchase Value"] = purchase_analysis_df["Total Purchase Value"].map("${:.2f}".format)
purchase_analysis_df["Avg Total Purchase per Person"] = purchase_analysis_df["Avg Total Purchase per Person"].map("${:.2f}".format)

# Display
purchase_analysis_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
age_bins = [0,9.99,14.99,19.99,24.99,29.99,34.99,39.99,100]
age_bin_names = ["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]

purchase_data_with_age_bin_unique = purchase_data.drop_duplicates(subset=["SN"])
player_num_total = purchase_data_with_age_bin_unique["SN"].count()
purchase_data_with_age_bin_unique["Age Range"] = pd.cut(purchase_data_with_age_bin_unique["Age"],age_bins,labels = age_bin_names)
player_num_by_age = purchase_data_with_age_bin_unique.groupby(["Age Range"])["SN"].count()
                       
player_percent_by_age = player_num_by_age / player_num_total

# Analysis Data Frame
age_demographics_analysis_df = pd.concat([player_num_by_age,player_percent_by_age],axis = 1)
age_demographics_analysis_df.columns = ["Total Count","Percentage of Players"]

# Display
age_demographics_analysis_df
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# Group by age bin
age_bins = [0,9.99,14.99,19.99,24.99,29.99,34.99,39.99,100]
age_bin_names = ["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]

purchase_data_with_age_bin_duplicate = purchase_data
purchase_data_with_age_bin_duplicate["Age Range"] = pd.cut(purchase_data_with_age_bin_duplicate["Age"],age_bins,labels = age_bin_names)

purchase_data_group_by_age_bin_duplicate = purchase_data_with_age_bin_duplicate.groupby(["Age Range"])

# Analysis calculation
purchase_count_by_age_bin = purchase_data_group_by_age_bin_duplicate["Purchase ID"].count()
avg_purchase_price_by_age_bin = round(purchase_data_group_by_age_bin_duplicate["Price"].mean(),2)
total_purchase_price_by_age_bin = round(purchase_data_group_by_age_bin_duplicate["Price"].sum(),2)
avg_purchase_price_per_person_by_age_bin = round(total_purchase_price_by_age_bin / purchase_data_with_age_bin_unique.groupby(["Age Range"])["SN"].count(),2)

# Analysis Data Frame
purchase_analysis_by_age_df = pd.concat([purchase_count_by_age_bin,avg_purchase_price_by_age_bin,
                                        total_purchase_price_by_age_bin,avg_purchase_price_per_person_by_age_bin],axis = 1)
purchase_analysis_by_age_df.columns = ["Purchase Count","Average Purchase Price","Total Purchase Value","Avg Total Purchase per Person"]

# Change to dollar format
purchase_analysis_by_age_df["Average Purchase Price"] = purchase_analysis_by_age_df["Average Purchase Price"].map("${:.2f}".format)
purchase_analysis_by_age_df["Total Purchase Value"] = purchase_analysis_by_age_df["Total Purchase Value"].map("${:.2f}".format)
purchase_analysis_by_age_df["Avg Total Purchase per Person"] = purchase_analysis_by_age_df["Avg Total Purchase per Person"].map("${:.2f}".format)

# Display
purchase_analysis_by_age_df
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Group the purchase data by player name
purchase_data_group_by_player = purchase_data.groupby(["SN"])

# Calcute stats by player
player_purchase_count = purchase_data_group_by_player["Purchase ID"].count()
player_avg_purchase_price = round(purchase_data_group_by_player["Price"].mean(),2)
player_total_purchase_price = purchase_data_group_by_player["Price"].sum()

# Agg analysis data into data frame
purchase_analysis_by_player = pd.concat([player_purchase_count,player_avg_purchase_price,player_total_purchase_price],axis=1)
purchase_analysis_by_player.columns = ["Purchase Count","Average Purchase Price","Total Purchase Value"]

# Sort data
purchase_analysis_by_player.sort_values(by=["Total Purchase Value"],ascending=False,inplace=True)

# Dollar Format
purchase_analysis_by_player["Average Purchase Price"] = purchase_analysis_by_player["Average Purchase Price"].map("${:.2f}".format)
purchase_analysis_by_player["Total Purchase Value"] = purchase_analysis_by_player["Total Purchase Value"].map("${:.2f}".format)

# Display
purchase_analysis_by_player.head(5)
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Group the purchase data by Item ID
purchase_data_by_item = purchase_data.groupby(["Item ID","Item Name"])

# Calculate stats for each Item
purchase_count_by_item = purchase_data_by_item["Purchase ID"].count()
item_price_by_item = purchase_data_by_item["Price"].first()
total_purchase_value_by_item = purchase_data_by_item["Price"].sum()

# Agg analysis data into data frame
purchase_analysis_by_item = pd.concat([purchase_count_by_item,item_price_by_item,total_purchase_value_by_item],axis=1)
purchase_analysis_by_item.columns = ["Purchase Count","Item Price","Total Purchase Value"]

# Sort by purchase count
purchase_analysis_by_item_sort_count = purchase_analysis_by_item.sort_values(by=["Purchase Count"],ascending=False)

# Change to dollar format
purchase_analysis_by_item_sort_count["Item Price"] = purchase_analysis_by_item_sort_count["Item Price"].map("${:.2f}".format)
purchase_analysis_by_item_sort_count["Total Purchase Value"] = purchase_analysis_by_item_sort_count["Total Purchase Value"].map("${:.2f}".format)

# Display
purchase_analysis_by_item_sort_count.head()
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
# Sort by total purchase value
purchase_analysis_by_item_sort_total_value = purchase_analysis_by_item.sort_values(by=["Total Purchase Value"],ascending=False)

# Change to dollar format
purchase_analysis_by_item_sort_total_value["Item Price"] = purchase_analysis_by_item_sort_total_value["Item Price"].map("${:.2f}".format)
purchase_analysis_by_item_sort_total_value["Total Purchase Value"] = purchase_analysis_by_item_sort_total_value["Total Purchase Value"].map("${:.2f}".format)

# Display
purchase_analysis_by_item_sort_total_value.head()
```
